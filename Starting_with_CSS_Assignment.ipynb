1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in thec project where required.

File: index.html
```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>CSS Exercise</title>

    <!-- Internal CSS (Will apply unless overridden by Inline CSS) -->
    <style>
      .internal-box {
        width: 200px;
        height: 100px;
        background-color: blue;
      }
      .box {
        width: 200px;
        height: 100px;
        background-color: blue;
        color: white;
        text-align: center;
        line-height: 100px;
      }
    </style>

    <!-- Linking External CSS -->
    <link rel="stylesheet" href="styles.css" />
  </head>
  <body>
    <!-- Different Ways to Apply Css -->
    <!-- 1. Inline -->
    <div style="background-color: red; width: 200px; height: 100px">
      Inline CSS
    </div>
    <!-- 2. Internal CSS -->
    <div class="internal-box">Internal CSS</div>
    <!-- 3. External CSS -->
    <div class="external-box">External CSS</div>

    <!-- Priority Check -->
    <!-- Inline CSS (Has the highest priority) -->
    <div class="box" style="background-color: red">Inline CSS</div>
    <div class="box">Internal CSS</div>
    <div class="box">External CSS</div>
  </body>
</html>
```

File: styles.css
```
.external-box {
  width: 200px;
  height: 100px;
  background-color: green;
}

/* External CSS (Lowest priority among these three) */
.box {
  background-color: green;
}

```


2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>BEM Naming Example</title>

    <style>
      /* Base styles for the paragraph block */
      .content__paragraph {
        font-size: 16px;
        padding: 10px;
        margin: 10px 0;
      }

      /* Modifier for highlighted text */
      .content__paragraph--highlight {
        background-color: yellow;
        font-weight: bold;
      }

      /* Modifier for informational text */
      .content__paragraph--info {
        background-color: lightblue;
        color: darkblue;
      }

      /* Modifier for warning text */
      .content__paragraph--warning {
        background-color: pink;
        color: red;
        font-weight: bold;
      }
    </style>
  </head>
  <body>
    <div class="content">
      <p class="content__paragraph content__paragraph--highlight">
        This is a highlighted paragraph.
      </p>
      <p class="content__paragraph content__paragraph--info">
        This paragraph provides informational text.
      </p>
      <p class="content__paragraph content__paragraph--warning">
        Warning! This paragraph has a warning message.
      </p>
    </div>
  </body>
</html>
```

3. Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

File: index.html
```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Styled Form Example</title>

    <!-- Link to External CSS -->
    <link rel="stylesheet" href="styles.css" />
  </head>
  <body>
    <div class="form-container">
      <h2>Contact Us</h2>

      <form action="/submit" method="post">
        <div class="form-section">
          <label for="name">Name:</label>
          <input
            type="text"
            id="name"
            name="name"
            placeholder="Enter your name"
          />

          <label for="email">Email:</label>
          <input
            type="email"
            id="email"
            name="email"
            placeholder="Enter your email"
          />
        </div>

        <label for="message">Message:</label>
        <textarea
          id="message"
          name="message"
          rows="4"
          placeholder="Type your message"
        ></textarea>

        <button type="submit">Submit</button>
      </form>
    </div>
  </body>
</html>

```

File: styles.css
```
/* Custom color palette */
:root {
  --primary-bg: #f4f4f9;
  --form-bg: #ffffff;
  --input-bg: #e0f7fa;
  --button-bg: #00796b;
  --button-hover: #004d40;
  --text-color: #333;
}

* {
  box-sizing: border-box;
}

/* General form styling */
body {
  font-family: Arial, sans-serif;
  background-color: var(--primary-bg);
  display: flex;
  justify-content: center;
  align-items: center;
  height: 100vh;
  margin: 0;
}

.form-container {
  background: var(--form-bg);
  padding: 20px;
  border-radius: 8px;
  box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2);
  width: 360px;
  text-align: center;
}

/* Apply opacity to the form section */
.form-section {
  background-color: rgba(0, 121, 107, 0.2); /* 20% transparent */
  padding: 15px;
  border-radius: 5px;
  margin-bottom: 15px;
}

label {
  display: block;
  margin-top: 10px;
  font-weight: bold;
  color: var(--text-color);
}

input,
textarea {
  width: 100%;
  padding: 8px;
  margin-top: 5px;
  border: 1px solid #ccc;
  border-radius: 5px;
  background-color: var(--input-bg);
}

button {
  width: 100%;
  background-color: var(--button-bg);
  color: white;
  padding: 10px;
  border: none;
  border-radius: 5px;
  font-size: 16px;
  cursor: pointer;
  margin-top: 10px;
}

button:hover {
  background-color: var(--button-hover);
}
```
